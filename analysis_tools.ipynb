```python
_ftod_r = re.compile(
    br'^(-?)([0-9]*)(?:\.([0-9]*))?(?:[eE]([+-][0-9]+))?$')
def ftod(f):
    """Print a floating-point number in the format expected by PDF:
    as short as possible, no exponential notation."""
    s = str(f)
    m = _ftod_r.match(s)
    if not m:
        raise RuntimeError("unexpected floating point number format: {!a}"
                           .format(s))
    sign = m.group(1)
    intpart = m.group(2)
    fractpart = m.group(3)
    exponent = m.group(4)
    if ((intpart is None or intpart == b'') and
        (fractpart is None or fractpart == b'')):
        raise RuntimeError("unexpected floating point number format: {!a}"
                           .format(s))

    # strip leading and trailing zeros
    if intpart is None: intpart = b''
    else: intpart = intpart.lstrip(b'0')
    if fractpart is None: fractpart = b''
    else: fractpart = fractpart.rstrip(b'0')

    if intpart == b'' and fractpart == b'':
        # zero or negative zero; negative zero is not useful in PDF
        # we can ignore the exponent in this case
        return b'0'

    # convert exponent to a decimal point shift
    elif exponent is not None:
        exponent = int(exponent)
        exponent += len(intpart)
        digits = intpart + fractpart
        if exponent <= 0:
            return sign + b'.' + b'0'*(-exponent) + digits
        elif exponent >= len(digits):
            return sign + digits + b'0'*(exponent - len(digits))
        else:
            return sign + digits[:exponent] + b'.' + digits[exponent:]

    # no exponent, just reassemble the number
    elif fractpart == b'':
        return sign + intpart # no need for trailing dot
    else:
        return sign + intpart + b'.' + fractpart
    
def format_fancy_interval(val_str, interval_str, units):
    #post processing on value +/- interval, with units. surrounds in scientific notation if better
    if re.search(r"(\d+)e(-\d+)", val_str) or re.search(r"(\d+)e(-\d+)", interval_str):
        val = float(val_str)
        interval = float(interval_str)
        #found scientific notation, surround the interval with parentheses and 
        #suffix the scientific notation part afterwards
        (_, interval_first_digit) = round_CI(interval)
        #interval_first_digit += 1
        val = ftod(val * 10**interval_first_digit)
        interval = ftod(interval * 10**interval_first_digit)
        interval = fix_scientific_stripped_zero(interval)
        return (r"(%s\pm{%s})e%d %s") % (val, interval, int(interval_first_digit * -1), units)
    else:
        return "(%s\pm{%s}) %s" % (val_str, interval_str, units)
```

```python
def print_unit_division(numerator_units, denominator_units):
    if numerator_units and len(numerator_units):
        if denominator_units and len(denominator_units):
            #have both numerator and denominator
            return r"\frac{" + numerator_units + "}{" + denominator_units + "}"
        else:
            #only have numerator
            return r"{" + numerator_units + "}"
    elif denominator_units and len(denominator_units):
        #only have denominator
        return r"{" + denominator_units + "}^{-1}"
    else:
        #dont have anything
        return ""
```

```python
def student_t(dof, confidence_level):
    exact_val = sc.stats.t.ppf(df=dof, q=(1+confidence_level)/2)
    rounded = np.round(exact_val, 3)
    return rounded
```

```python
def presentation_xbar(df, interval_decimal_place):
    mean = df.mean()
    str_representation = str(mean)
    #if standard deviation is zero then dont do rounding magic
    if df.drop_duplicates().count() > 1:
        str_representation = round_xbar(mean, interval_decimal_place)
    #if "." not in str_representation:
    #    str_representation = str_representation + "."
    return str_representation

def round_xbar(mean, interval_decimal_place):
    first_digit_decimals = interval_decimal_place + 1
    rounded_next_to_decimal = np.round(mean * np.power(10, first_digit_decimals))
    xbar_rounded = rounded_next_to_decimal * np.power(10, -first_digit_decimals)
    if interval_decimal_place < 0:
        digits_after_decimal = 0
    else:
        digits_after_decimal = int(interval_decimal_place)
    str_representation = ("%." + str(digits_after_decimal) + "f") % xbar_rounded
    return str_representation
    
def presentation_stddev(df, digits = 4):
    if df.drop_duplicates().count() == 1:
        stddev = 0
        first_digit_decimals = -1
    else:
        stddev = df.std()
        first_digit_decimals = np.floor(-1 * np.log10(stddev))
    rounded_next_to_decimal = np.round(stddev * np.power(10, first_digit_decimals), digits)
    stddev_rounded = rounded_next_to_decimal * np.power(10, -first_digit_decimals)
    digits_before_decimal = 0 if first_digit_decimals > 0 else int(-1 * first_digit_decimals)
    digits_after_decimal = 0 if digits + int(first_digit_decimals) < 0 else digits + int(first_digit_decimals)
    str_representation = ("%" + str(digits_before_decimal) + "." + str(digits_after_decimal) + "f") % stddev_rounded
    if "." not in str_representation:
        str_representation = str_representation + "."
    return str_representation
          
def presentation_confidence_interval(df, level = .95, round_answer = True):
    n = df.count()
    s = df.std()
    dof = n - 1
    t_factor = student_t(dof, level)
    CI = t_factor * s / np.sqrt(n)
    str_representation = str(CI)
    first_digit_decimals = 0
    if df.drop_duplicates().count() == 1:
        #all items are the same, standard deviation = 0
        str_representation = "0"
        first_digit_decimals = 0
    elif round_answer:
        (str_representation, first_digit_decimals) = round_CI(CI)
    return (str_representation, first_digit_decimals)

def fix_scientific_stripped_zero(str_val):
    str_val = re.sub(r"^(-?)([12](?=[e$]))(e?)((-?\d+)?)$", r"\g<1>\g<2>.0\g<3>\g<4>", str_val)
    str_val = re.sub(r"^(-?)0\.([12])(e)((-\d+))$", r"\g<1>.\g<2>0\g<3>\g<4>", str_val)
    str_val = re.sub(r"^(-?)([12])$", r"\g<1>\g<2>.0", str_val)
    str_val = re.sub(r"^(-?)0\.([12])$", r"\g<1>0.\g<2>0", str_val)
    return str_val

def round_CI(CI):
    first_digit_decimals = -1 * np.floor(np.log10(CI))
    first_digit = str(CI * np.power(10, first_digit_decimals))[0]
    digits_to_use = 2 if int(first_digit) < 3 else 1
    next_to_decimal = CI * np.power(10, first_digit_decimals - 1)
    rounded_next_to_decimal = np.round(next_to_decimal, digits_to_use)
    CI_rounded = rounded_next_to_decimal * np.power(10, -first_digit_decimals + 1)
    #check first digit again, in case we rounded up and lost precision
    first_digit_decimals = -1 * np.floor(np.log10(CI_rounded))
    str_representation = "%g" % CI_rounded
    str_representation = fix_scientific_stripped_zero(str_representation)
    str_representation = str_representation.strip()
    #if "." not in str_representation:
    #    str_representation = str_representation + "."
    return (str_representation, first_digit_decimals)

def do_error_analysis(data, key_column, val_column, vals_units, confidence_level):
    vals_by_group = data.groupby(key_column)
    group_names = vals_by_group.groups.keys()
    presentation_data = pd.DataFrame(index=pd.Index(data=group_names, name=key_column), columns=["xbar", "interval", "interval_rounded", "t"])
    for group_name in group_names:
        group_vals = vals_by_group.get_group(group_name)[val_column]
        (interval_rounded, interval_first_digit) = presentation_confidence_interval(group_vals, level=confidence_level)
        (interval, _) = presentation_confidence_interval(group_vals, level=confidence_level, round_answer=False)
        xbar = presentation_xbar(group_vals, interval_first_digit)
        presentation_data.ix[group_name].xbar = xbar
        presentation_data.ix[group_name].interval = interval
        presentation_data.ix[group_name].interval_rounded = interval_rounded
        presentation_data.ix[group_name].t = student_t(group_vals.count()-1, confidence_level)
    stddevs = vals_by_group[val_column].apply(presentation_stddev)
    presentation_data["standard deviation"] = stddevs
    presentation_data["err_analysis"] = presentation_data.apply(lambda row: "(" + row.xbar + " ± " + row.interval_rounded + ") " + vals_units, axis=1)
    
    #print equations out for the report
    for group_name in group_names:
        show_big_text(group_name)
        df = presentation_data.ix[group_name]
        group_vals = vals_by_group.get_group(group_name)[val_column]
        n = group_vals.count()
        render_math_image(r"n=" + str(n))
        render_math_image(r"\bar{x}=" + df.xbar +  " " + vals_units + r" from software")
        s = presentation_stddev(group_vals)
        render_math_image(r"s=" + s + " " + vals_units + r" from software")
        t_factor = np.round(student_t(n-1, confidence_level), 3).astype(str)
        render_math_image(r"t_{" + str(int(100*confidence_level)) + "}=" + t_factor + ", " + str(n-1) + " degrees of freedom, from software")
        render_math_image(r"\mu_{" + str(int(100*confidence_level)) + "}=(\bar{x}\pm\frac{s\times{t_{" + str(int(100*confidence_level)) + "}}}{\sqrt{n}}) " + vals_units)
        render_math_image(r"=(" + df.xbar + r"\pm\frac{" + s + r"\times{" + t_factor + r"}}{\sqrt{" + str(n) + r"}}) " + vals_units)
        render_math_image(r"=") + format_fancy_interval(df.xbar, dv.interval_rounded, vals_units)
    return presentation_data
```

```python
def do_linear_regression(m, m_stderr, b, b_stderr, y_units, x_units, n, confidence_level, show_steps = True):
    #this ones easy
    
    #print what excel/google sheets gave us
    y_over_x_units = print_unit_division(y_units, x_units)
    if show_steps:
        render_math_image(r"From spreadsheet linest( ) function...")
        render_math_image(r"m=" + str(m) + r" " + y_over_x_units)
        render_math_image(r"s_m=" + str(m_stderr) + r" " + y_over_x_units)
        render_math_image(r"b=" + str(b) + r" {" + y_units + "}")
        render_math_image(r"s_b=" + str(b_stderr) + r" {" + y_units + "}")
        render_math_image(r"\frac{                                                         }{                                                         }")
    #start setting up the error analysis
    n=int(n)
    if show_steps:
        render_math_image(r"n=" + str(n))
    dof_variables = 2
    dof = n - dof_variables
    if show_steps:
        render_math_image(r"DOF=" + str(n) + "-" + str(dof_variables) + "=" + str(dof))
    t = round(student_t(dof, confidence_level), 3)
    if show_steps:
        render_math_image(r"t_{" + str(int(100*confidence_level)) + "}=" + str(t))
    #m with confidence interval
    if show_steps:
        render_math_image(r"m_{" + str(int(100*confidence_level)) + r"}=(" + str(m) + r"\pm\frac{" + str(m_stderr) + r"\times{" + str(t) + r"}}{\sqrt{" + str(n) + r"}}) " + y_over_x_units)
    m_ci_unrounded = m_stderr * t / np.sqrt(n)
    m_ci_rounded_data = round_CI(m_ci_unrounded)
    m_ci_rounded = m_ci_rounded_data[0]
    m_ci_first_digit = m_ci_rounded_data[1]
    m_rounded = round_xbar(m, m_ci_first_digit)
    if show_steps:
        render_math_image(r"=(" + str(m_rounded) + r"\pm" + m_ci_rounded + r") " + y_over_x_units)
    #b with confidence interval
    if show_steps:
        render_math_image(r"b_{" + str(int(100*confidence_level)) + r"}=(" + str(b) + r"\pm\frac{" + str(b_stderr) + r"\times{" + str(t) + r"}}{\sqrt{" + str(n) + r"}}) {" + y_units + r"}")
    b_ci_unrounded = b_stderr * t / np.sqrt(n)
    b_ci_rounded_data = round_CI(b_ci_unrounded)
    b_ci_rounded = b_ci_rounded_data[0]
    b_ci_first_digit = b_ci_rounded_data[1]
    b_rounded = round_xbar(b, b_ci_first_digit)
    if show_steps:
        render_math_image(r"=(" + str(b_rounded) + r"\pm" + b_ci_rounded + r") {" + y_units + r"}")
    render_math_image(r"y_{" + str(int(100*confidence_level)) + r"}=((" + str(m_rounded) + r"\pm" + m_ci_rounded + r")x+(" + str(b_rounded) + r"\pm" + b_ci_rounded + r")) {" + y_units + r"}")
    return pd.Series(data={
                           "m": m,
                           "m_rounded": m_rounded,
                           "m_ci": m_ci_unrounded,
                           "m_ci_rounded": m_ci_rounded,
                           "b": b,
                           "b_rounded": b_rounded,
                           "b_ci": b_ci_unrounded,
                           "b_ci_rounded": b_ci_rounded,
                           "t": t,
                           })
```

# error propogation

```python
#todo: need a way to easily pass in unrounded confidence interval
def present_error_propogation(symbols, equation, equation_name, equation_units, show_steps = True):
    symbol_value_substitutions = {symbol.ref: symbol.value for symbol in symbols}
    if show_steps:
        render_math_image(equation_name + r"=(" + sympy.latex(equation) + r"\pm\lambda_{" + equation_name + r"}) " + equation_units)
    line_1 = []
    line_2 = []
    line_3 = []
    
    #get confidence interval
    grand_lambda_squared = 0
    for symbol in symbols:
        if (symbol.include_in_error_calcs):
            partial_derivative = equation.diff(symbol.ref)
            line_1.append(r'(\frac{\partial ' + equation_name + r'}{\partial ' + symbol.expression + r'})^{2}\lambda_{' + symbol.expression + r'}^{2}')
            line_2.append(r'(' + sympy.latex(partial_derivative) + r')^{2}\lambda_{' + symbol.expression + r'}^{2}')
            tmp_equation = partial_derivative
            for symbol_2 in symbols:
                tmp_equation = tmp_equation.replace(symbol_2.ref, sympy.Symbol("(" + str(symbol_2.value) + str(symbol_2.units) + r")"))
            line_3.append(sympy.latex(tmp_equation ** 2) + r"\times" + sympy.latex(sympy.Symbol("(" + str(symbol.confidence_interval) + str(symbol.units) + r")") **2))
            grand_lambda_squared += partial_derivative.evalf(subs=symbol_value_substitutions) **2 * symbol.confidence_interval **2
    if show_steps:
        render_math_image(r"\lambda_" + equation_name + r"=\sqrt{" + "+".join(line_1) + "}")
        render_math_image(r"=\sqrt{" + "+".join(line_2) + "}")
        render_math_image(r"=\sqrt{" + "+".join(line_3) + "}")
    grand_lambda = float(str(sympy.sqrt(grand_lambda_squared)))
    ci_data = round_CI(grand_lambda)
    grand_lambda_rounded = ci_data[0]
    interval_decimal_place = ci_data[1]
    if show_steps:
        render_math_image(r"=" + grand_lambda_rounded + equation_units)
    
    #get value
    tmp_equation = equation
    for symbol in symbols:
        tmp_equation = tmp_equation.replace(symbol.ref, sympy.Symbol("(" + str(symbol.value) + str(symbol.units) + r")"))
    if show_steps:
        render_math_image(sympy.latex(equation) + r"=" + sympy.latex(tmp_equation))
    grand_val = float(str(equation.evalf(subs=symbol_value_substitutions)))
    grand_val_rounded = round_xbar(grand_val, interval_decimal_place)
    if show_steps:
        grand_val_rounded_sci_notation_optional = ("%g" % float(grand_val_rounded))
        render_math_image(r"=" + grand_val_rounded_sci_notation_optional + equation_units)
    
    #print whole expression
    render_math_image(equation_name + r"=" + format_fancy_interval(grand_val_rounded, grand_lambda_rounded, equation_units))
    
    return (grand_val, grand_lambda)

class define_symbol:
    def __init__(self, expression, value, confidence_interval = None, units = None, include_in_error_calcs = True, round_decimal_digits = None):
        self.expression = expression
        self.value = np.float(value)
        self.confidence_interval = np.float(confidence_interval)
        if round_decimal_digits is not None:
            (self.value, self.confidence_interval) = np.round([self.value, self.confidence_interval], round_decimal_digits)
        self.units = units
        self.include_in_error_calcs = include_in_error_calcs
        self.ref = sympy.Symbol(expression)
```
